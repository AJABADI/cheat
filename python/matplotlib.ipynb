# Simple examples

```python
x = np.linspace(0, 10)
# plt.axis([0, 5, 0, 100]) # like plt.xlim(), plt.ylim()
plt.xlim(0, 5)
plt.ylim(0, 100)
lines = plt.plot(x, x, 'r', x, x**2, 'b', x, x**3, 'g')
for line in lines:
    plt.setp(line, linewidth=2)
plt.text(2, 30, '$f(x)=x^3$', fontsize=18)
plt.annotate('$f(x)=x^2$', xy=(4, 20), xytext=(4.5, 40), fontsize=18, arrowprops=dict(facecolor='black', shrink=0.05))
plt.grid()
plt.title('Polynomials')
plt.xlabel('x')
plt.ylabel('y')
```

```python
plt.figure() # optional, plt.plot() creates first figure be default
plt.subplot(2, 1, 1) # subplot(1, 1, 1) is optional
x = np.linspace(0, 5) 
plt.plot(x, np.sin(x), 'r', linewidth=2)
plt.axes([0.5, 0.6, 0.2, 0.2]) # add additional axes; [l, b, w, h]
plt.plot(x, np.sin(x), 'g', linewidth=2)

plt.subplot(212) # commas can be ignored
plt.plot(x, np.cos(x), 'b', linewidth=2)
```

# Annotations

 * http://matplotlib.org/users/annotations_guide.html

```python
plt.figure()
ax = plt.subplot()
x = np.linspace(-1, 1)
kernel = lambda x: np.maximum(1 - x**2, 0)
plt.plot(x, kernel(x), linewidth=2)
plt.axis([-1.5, 1.5, 0, 1.5])
bbox_props = dict(boxstyle='square', linewidth=2, facecolor='red', edgecolor='blue', alpha=0.5)
bbox_props = dict(boxstyle='rarrow', lw=2, fc='red', ec='blue', alpha=0.5)
t = plt.text(0, 1.3, 'Maximum', bbox=bbox_props, ha='right', fontsize=12, rotation=-45)
bbox = t.get_bbox_patch()
bbox.set_facecolor('grey')
plt.scatter(0.5, kernel(0.5), color='red', s=25, marker='s')
plt.annotate('Point', xy=(0.5, kernel(0.5)), xytext=(0.0, 0.4), size=18,  
             bbox=dict(boxstyle='round', ec='black', fc='grey'), 
             arrowprops=dict(arrowstyle='-|>', lw=2))
plt.annotate('Top right', xy=(1.0, 1.0), xytext=(1.0, 1.0), xycoords='axes fraction',
             va='top', ha='right', size=20, color='red')
```

## Colormaps and mappers

```python
x = np.linspace(-5, 5, 40)
y = stats.norm(loc=0.0, scale=1.0).pdf(x)
fig, ax = plt.subplots()

# Normalizes [vmin, vmax] -> [0.0, 1.0]
normalize = colors.Normalize(vmin=np.min(y), vmax=np.max(y))
# Colormap maps values [0;1] -> [colormin; colormax]
cmap = cmx.hot_r
# ScalarMappable uses norm and cmap to map values [vmin; vmax] -> [colormin; colormax]
mapper = cmx.ScalarMappable(norm=normalize, cmap=cmap)
color = mapper.to_rgba(y)

ax.bar(x, y, width=(np.max(x) - np.min(x))/len(x), color=color)
```

# Linestyles and marker styles

```python
fig, axes = plt.subplots(1, 2, figsize=(10, 3))
groups = [['-', '--', ':', '-.'], ['solid', 'dashed', 'dotted', 'dashdot']]
markers = ['s', 'o', 'd', 'x']
x = range(0, 10)
for i in range(len(groups)):
    for j, linestyle in enumerate(groups[i]):
        line, = axes[i].plot(x, [j + 1] * len(x))
        line.set_linestyle(linestyle)
        line.set_linewidth(j + 1)
        line.set_marker(markers[j])
        line.set_markersize(10)
    axes[i].set_ylim(0, len(groups[i]) + 1)
```

# Animations

```python
from matplotlib import animation
```

```python
fig = plt.figure()
x = np.linspace(0, 10)
ax = plt.axes(xlim=(0, max(x)), ylim=(-2, 2))
sine, = ax.plot([], [])

# Will be called when an FuncAnimation object is created
def ani_init():
    sine.set_data([], [])
    sine.set_linewidth(2)
    sine.set_color('blue')
    return sine

# i: frame number
def ani_update(i):
    sine.set_data(x, 2*np.sin(x + i))
    return sine

# Sets up animation
# calls init_func; does not yet produce the animation
# init_func is optional
# blit: only update parts of the frame -> smoother animation
ani = animation.FuncAnimation(fig, ani_update, init_func=ani_init, frames=20, blit=True)
writer = animation.FFMpegWriter(fps=5) # fps must be at least 5!
# produces the animation; length in sec=frames/fps
ani.save('sine.mp4', extra_args=['-vcodec', 'libx264'], writer=writer)
    
```
