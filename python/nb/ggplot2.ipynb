```python
import ggplot
import matplotlib
from ggplot import *
from ggplot import diamonds
```

```python
%matplotlib inline
```

```python
data = diamonds[:1000]
```

```python
p = ggplot(data, aes(x='x', y='y', color='color')) + geom_point() + xlab('X-value') + ylab('Y-value')
```

```python
print(p)
```

```python
p + facet_wrap('cut')
```

```python
ggplot(data, aes(x='x')) + geom_histogram(binwidth=0.3, fill='lightblue', color='black')
```

```python
ggplot(data, aes(x='x')) + geom_bar(stat='bin')
```
