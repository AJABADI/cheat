# Python2R

```python
from rpy2.robjects import r # for assigning variables and executing commands
```

```python
def save_r(name, filename):
    r('save({:s}, file="{:s}")'.format(name, filename))
```

## Saving numpy arrays

```python
import rpy2.robjects.numpy2ri # activate conversion numpy -> RData
```

```python
mat = np.random.rand(3, 4)
```

```python
r.assign('mat', mat);
save_r('mat', 'mat.RData')
```

## Saving pandas arrays
